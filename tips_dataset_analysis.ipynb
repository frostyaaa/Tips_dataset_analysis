```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size", palette="deep",)
plt.title("scatterplot of total bill vs tip")
plt.show()
```

```python
sns.lineplot(data=tips, x='size', y='total_bill', hue='sex', markers='o',ci=None)
plt.title("lineplot for total_bill vs size")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", palette="muted")
plt.title("barplot of dat vs total_bill")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="tip",hue="smoker", palette="dark")
plt.title("boxplot for tips by day and smoker status")
plt.show()
```

```python
sns.violinplot(data=tips,x='day', y='total_bill',hue='time', palette="dark", split=True)
plt.title("violin plot of total_bill by day and time")
plt.show()
```

```python
sns.countplot(data=tips,x='day',palette='dark',hue="smoker")
plt.title("count plot")
plt.show()
```

```python
sns.regplot(data=tips,x='total_bill',y='tip',scatter_kws={'s':50},line_kws={'color':'red'})
plt.title("regression plot of total_bill")
plt.show()
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,color='blue', bins=20)
plt.title("histplot of total_bill")
plt.show()
```

```python
sns.pairplot(tips, hue='sex', vars=['total_bill', 'tip', 'size'], palette='dark')
plt.title("pairplot:numeric variables by gender",y=1.02)
plt.show()
```

```python
sns.catplot(data=tips,x='day', y='tip',hue='sex',kind='point',palette='bright')
plt.title("catplot for day vs tip")
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker', palette='dark')
plt.suptitle("joinplot for tip by smoker")
plt.show()
```

```python
g= sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill',bins=20, kde=True)
g
```
